### 6)create a list from rows from dataframe

```python
rows_list=[]
for index,rows in df.iterrows():
    temp_list=[rows.names,rows.age,rows.place]
    rows_list.append(temp_list)
print(rows_list)
```

### 7)create a dictionary by taking only two columns of dataframe

```python
dictionary_of_result={}
list_of_names=list(df["names"])
# print(type(list_of_names))
list_of_places=list(df["place"])
# print(type(list_of_places))
for i in range(len(list_of_names)):
    dictionary_of_result[list_of_names[i]]=list_of_places[i]
print(dictionary_of_result)
```
